# Relationship between variables using Matplotlib graphs

```python
#Import libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#Import the dataset
car = pd.read_csv("C:/Users/a/Himanshu/Dataset/Cars93.csv")
```

```python
car.columns
```

```python
#Creating empty list for categorical as well as continous variable
CAT = []
CON = []
```

```python
#Divide the categorical and continous variable in two different list
for i in car.columns:
    if (car[i].dtype == 'object'):
        CAT.append(i)
    else:
        CON.append(i)
```

```python
CAT
```

```python
CON
```

```python
#Univariate Graph
car['AirBags'].value_counts().plot(kind="barh")
```

```python
#Bivariate Graph
for i in CON:
    sb.boxplot(car['AirBags'],car[i])
    plt.show()
```

```python
#Bivariate graph of two categorical variable
for i in CAT:
   print(pd.crosstab(car['AirBags'],car[i],normalize=True))
```
